```python
import random, time
import matplotlib.pyplot as plt

temps = []
hums = []

for t in range(10):
    temp = 25 + random.uniform(-1, 1)
    hum = 60 + random.uniform(-5, 5)
    temps.append(temp)
    hums.append(hum)

    with open("data.txt", "a") as f:
        f.write(f"{t},{temp},{hum}\n")

    time.sleep(0.5)

plt.plot(temps)
plt.xlabel("Time (s)")
plt.ylabel("Temperature (C)")
plt.title("Temperature Log")
plt.show()

plt.plot(hums)
plt.xlabel("Time (s)")
plt.ylabel("Humidity (%)")
plt.title("Humidity Log")
plt.show()
```
